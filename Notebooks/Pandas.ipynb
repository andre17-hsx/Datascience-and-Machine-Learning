```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#### IMPUTACION
```

```python
oil = pd.read_csv("../data/oil.csv")
oil
```

```python
oil.dtypes
```

```python
oil["date"] = oil["date"].astype("datetime64[ns]")
```

```python
filtro2015 = (oil["date"] >= "2015-01-01") & (oil["date"] <= "2015-12-31")
filtro2016 = (oil["date"] >= "2016-01-01") & (oil["date"] <= "2016-12-31")
filtro2017 = (oil["date"] >= "2017-01-01") & (oil["date"] <= "2017-12-31")
aceite2015 = oil.loc[filtro2015]
aceite2015['dcoilwtico'].fillna(aceite2015['dcoilwtico'].mean(), inplace = True)
aceite2015['date'] = aceite2015['date'].dt.strftime('%d-%m')
aceite2015
```

```python
aceite2016 = oil.loc[filtro2016]
aceite2016['dcoilwtico'].fillna(aceite2016['dcoilwtico'].mean(), inplace = True)
aceite2016['date'] = aceite2016['date'].dt.strftime('%d-%m')
aceite2016
```

```python
aceite2017 = oil.loc[filtro2017]
aceite2017['dcoilwtico'].fillna(aceite2017['dcoilwtico'].mean(), inplace = True)
aceite2017['date'] = aceite2017['date'].dt.strftime('%d-%m')
aceite2017
```

```python
sbn.set_style("darkgrid")
fig,axs = plt.subplots(figsize=(8,8))
sbn.lineplot(data= aceite2015, x="date", y = "dcoilwtico", color="b",marker="o", linestyle="-.",ax=axs, label="2015")
sbn.lineplot(data= aceite2016, x="date", y = "dcoilwtico", color="r",marker="s", linestyle="--",ax=axs, label="2016")
sbn.lineplot(data= aceite2017, x="date", y = "dcoilwtico", color="g",marker="s", linestyle="-.",ax=axs, label="2017")
_=plt.xticks(rotation=90)
axs.set_xlabel("Fechas")
axs.set_ylabel("Precios")
axs.set_title("Precios de aceite en 2015, 2016 y 2017")
plt.savefig("lineplot.png")
```

Del gráfico se puede observar como en el año 2017 el precio del petróleo se mantuvo estable, teniendo su precio entre 45 y 55.
En el año 2016 se puede observar como los primeros meses del año el petroleo tuvo un precio considerablemente bajo, hasta que consiguio alcanzar el precio estandar de 45 a 55. Luego, tuvo una caida un poco pasada la mitad del año en donde bajo hasta 30 y volvió a subir en los últimos meses del año a sus precios regulares de 45 a 55.
En el año 2015 se puede observar como dentro del primer tercio del año el precio del petróleo subió hasta más de 60, teniendo una de las mayores cifras en comparación a todos los años y luego disminuyo llegando hasta 35, en este caso se ve que el precio sufrio una gran inestabilidad. 

```python
### ZSCORES
```

```python
aceite2015["zscores"] = stats.zscore(aceite2015["dcoilwtico"])
aceite2015
```

```python
aceite2016["zscores"] = stats.zscore(aceite2016["dcoilwtico"])
aceite2016
```

```python
aceite2017["zscores"] = stats.zscore(aceite2017["dcoilwtico"])
aceite2017
```

```python
sbn.set_style("darkgrid")
fig,axs = plt.subplots(1,3,figsize=(12,10))
_=sbn.distplot(aceite2017["zscores"],ax=axs[0],color="red").set(title='2017')
axs[0].set_ylim(ymin=0, ymax=0.55)
_=sbn.distplot(aceite2016["zscores"],ax=axs[1],color="blue").set(title='2016')
axs[1].set_ylim(ymin=0, ymax=0.55)
_=sbn.distplot(aceite2015["zscores"],ax=axs[2],color="green").set(title='2015')
axs[2].set_ylim(ymin=0, ymax=0.55)
plt.suptitle("DESVIACIONES DE LOS AÑOS 2015,2016,2017")
fig.tight_layout()
```

```python
##HEATMAPS
```

```python
dfene = pd.read_csv("../data/sales_2017_enero.csv")
dfene
```

```python
dffeb = pd.read_csv("../data/sales_2017_febrero.csv")
dffeb
```

```python
dfmar = pd.read_csv("../data/sales_2017_marzo.csv")
dfmar
```

```python
dfmar["date"] = dfmar["date"].astype("datetime64[ns]")
dffeb["date"] = dffeb["date"].astype("datetime64[ns]")
dfene["date"] = dfene["date"].astype("datetime64[ns]")
dfmar['diaSemana'] =  dfmar['date'].dt.day_name()
dffeb['diaSemana'] =  dffeb['date'].dt.day_name()
dfene['diaSemana'] =  dfene['date'].dt.day_name()
```

```python
dfmar
```

```python
stores = pd.read_csv("../data/stores.csv")
stores
```

```python
dfmar = dfmar.merge(stores,left_on = "store_nbr", right_on="store_nbr", how="left")
dffeb = dffeb.merge(stores,left_on = "store_nbr", right_on="store_nbr", how="left")
dfene = dfene.merge(stores,left_on = "store_nbr", right_on="store_nbr", how="left")
dfene
```

```python
dfeneNew= dfene.groupby(['diaSemana', 'state']).size().unstack(level=1)
del dfeneNew['Guayas']
del dfeneNew['Pichincha']
dfeneNew
```

```python
dffebNew= dffeb.groupby(['diaSemana', 'state']).size().unstack(level=1)
del dffebNew['Guayas']
del dffebNew['Pichincha']
dffebNew
```

```python
dfmarNew= dfmar.groupby(['diaSemana', 'state']).size().unstack(level=1)
del dfmarNew['Guayas']
del dfmarNew['Pichincha']
dfmarNew
```

```python
sbn.set_style("darkgrid")

fig,axs = plt.subplots(1,3,figsize=(12,12))

sbn.heatmap(dfeneNew, ax=axs[0])
axs[0].set_ylim(ymin=7500, ymax=3000)
axs[0].set_title('Enero')

sbn.heatmap(dffebNew, ax=axs[1])
axs[1].set_ylim(ymin=7500, ymax=3000)
axs[1].set_title('Febrero')

sbn.heatmap(dfmarNew, ax=axs[2])
axs[2].set_ylim(ymin=7500, ymax=3000)
axs[2].set_title('Marzo')

plt.suptitle("ANALISIS DE TRANSACCIONES POR PROVINCIA")
fig.tight_layout()
plt.show()
```

Se puede observar como en febrero se tuvo los  valores mas pequeños en ventas mientras que en enero y febrero se tuvo valores más regulares. Los días martes en la provincia de azuay parece ser que hay muchas transacciones
