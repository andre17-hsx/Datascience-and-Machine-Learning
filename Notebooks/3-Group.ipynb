```python
files= [
    "../Data/sales_2017_enero.csv",
    "../Data/sales_2017_febrero.csv",
    "../Data/sales_2017_marzo.csv"
]
```

```python
stores= pd.read_csv("../Data/stores.csv")
```

```python
stores.head()
```

```python
#Para llevar datas de un dataframe a otro, usamos merge, para ello debe haber una columna que sea igual en ambas
```

```python
df.merge(stores, on="store_nbr", how="inner")
```

```python
df.shape
```

```python
df= df.merge(stores, left_on="store_nbr", right_on= "store_nbr", how="inner") #si no existiecen columnas con el mismo nombre en ambis dataframes utilizo left_on para el dataframe de la izquierda y right_on para el de la derecha
```

```python
df.shape
```

```python
df.groupby("state").size().sort_values() #devuelve una serie cuyo indices son las provincias y en el valor, la cantidad de filas que tienen esa provincia
```

```python
df.groupby("state").size().idxmin() #primero agrupo por provincias, luego con .size() sumo el total de tiendas para cada provincia, y finalmente .idxmin() me devuleve el indice que tiene el menor valor
```

```python
df.groupby("state").size().idxmax() #me devuelve el indice que tiene el mayor valor
```

```python
df.groupby("state")["unit_sales"].sum().idxmax()  #Si me interesa una fila especifica despues de haber separado por provincias, lo pongo despues entre corchetes
```

```python
df.groupby("state")["store_nbr"].nunique()  #nunique() me da la cantidad de valores unicos.
```

```python
df.groupby(["state","city"])["unit_sales"].sum().reset_index(drop= False) #si le hago reset index con drop False, se transforma en dataframe cuyos indices anteriores pasan a ser columnas
```

```python
#uso de funcion agg para hacer varias operaciones en una sola, recibe la columna a la cual quiero sacar las distintas caracteristicas
```

```python
dfg = df.groupby("city").agg({"unit_sales":["min","max","mean","sum"],"item_nbr":"nunique"}) #agg devuelve un datrame
```

```python
dfg
```

```python
dfg.columns
```

```python
dfg= dfg.reset_index()
```

```python
dfg.columns = dfg.columns.droplevel() #con droplevel le quito un nivel a las columnas que eran tuplas
```
