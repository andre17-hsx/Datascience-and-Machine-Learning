```python
import  pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
dfene = pd.read_csv("sales_2017_enero.csv")
```

```python
dfene.head()
```

```python
stores = pd.read_csv("stores.csv")
```

```python
stores.head()
```

```python
dfm = dfene.merge(stores, on ="store_nbr", how = "left")
```

```python
dfm.head()
```

```python
dfm["date"] = pd.to_datetime(dfm["date"])
```

```python
dfg = dfm.groupby([pd.Grouper(key="date",freq="D",label="left"),"state"])["unit_sales"].sum().reset_index()
```

```python
dfg.head()
```

```python
dfgy = dfg[dfg["state"]=="Guayas"]
```

```python
dfgy.head()
```

```python
dfpi = dfg[dfg["state"]=="Pichincha"]
```

```python
dfpi.head()
```

```python
sbn.set_style("darkgrid")
fig,axs = plt.subplots(figsize=(8,8))
sbn.lineplot(data= dfgy, x="date", y = "unit_sales", color="b",marker="o", linestyle="-.",ax=axs, label="Guayas")
sbn.lineplot(data= dfpi, x="date", y = "unit_sales", color="r",marker="s", linestyle="--",ax=axs, label="Pichincha")
_=plt.xticks(rotation=90)
axs.set_xlabel("Fechas")
axs.set_ylabel("Unidades Vendidas")
axs.set_title("Unidades vendidas en Guayaquil y Quito durante Enero")
plt.savefig("lineplot.png")
```

```python
sbn.set_style("darkgrid")
fig,axs = plt.subplots(figsize=(8,8))
sbn.lineplot(data= dfg, x="date", y = "unit_sales", hue="state", ax= axs)
```

```python
dfh = dfm.groupby([pd.Grouper(key="date",freq="SMS",label="left"),"state"])["unit_sales"].sum().reset_index()
```

```python
dfh.head()
```

```python
dfh["date"].unique()
```

```python
vgy = dfh[dfh["state"]=="Guayas"]["unit_sales"].values
```

```python
vpi = dfh[dfh["state"]=="Pichincha"]["unit_sales"].values
```

```python
dfht = pd.DataFrame({"Guayas":vgy,"Pichincha":vpi})
```

```python
dfht.head()
```

```python
dfht.index = ["quincena1","quincena2"]
```

```python
dfht.head()
```

```python
dfht.index
```

```python
sbn.set_style("darkgrid")
fig,axs = plt.subplots(figsize=(5,5))
_=sbn.heatmap(dfht.iloc[:,:2], yticklabels=dfht.index, ax= axs, annot=True, fmt=".2f", cmap = "YlOrBr")
_=plt.yticks(rotation=0)
```

```python
sbn.set_style("darkgrid")
fig,axs = plt.subplots(2,3,figsize=(12,10))
sbn.lineplot(data= dfgy, x="date", y = "unit_sales", color="b",marker="o", linestyle="-.",ax=axs[0,0], label="Guayas")
axs[0,0].set_ylim(ymin=0, ymax=900000)
_=plt.setp(axs[0,0].get_xticklabels(), rotation=90)
sbn.lineplot(data= dfpi, x="date", y = "unit_sales", color="r",marker="s", linestyle="--",ax=axs[0,1], label="Pichincha")
axs[0,1].set_ylim(ymin=0, ymax=900000)
_=plt.setp(axs[0,1].get_xticklabels(), rotation=90)
plt.suptitle("ANALISIS DE UNIDADES VENDIDAS POR PROVINCIA")
fig.tight_layout()
```
