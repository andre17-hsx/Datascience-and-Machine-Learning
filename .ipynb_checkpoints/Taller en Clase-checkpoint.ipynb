```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("../data/data_2019-1.csv")
```

```python
df
```

```python
df.dtypes
```

```python
df.dropna(subset=["precio","metroscuadradostotales","latlong"],inplace=True)
```

```python
df
```

```python
df["precio"]= df["precio"].str.replace("$","").str.replace(".","",regex=False)
```

```python
df
```

```python
df["precio"].unique()
```

```python
df["precio"]= df["precio"].str.replace(" ","")
```

```python
df= df.loc[df["precio"]!="",:]
```

```python
df["precio"] = df["precio"].astype(float).astype(int)
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df["latlong"]
```

```python
df["latlong"].str.split("/", n=5, expand=True)[5].str.split(",",expand=True)#Con n tomo solamente hasta el indice que me interesa de la lista creada con split. y con expand=True hago que lo que devuelve sea un dataframe, adicional al final con [5] tomo solamente la columna de los valores de lat y long
```

```python
df[["lat","long"]] = df["latlong"].str.split("/", n=5, expand=True)[5].str.split(",",expand=True)#Con n tomo solamente hasta el indice que me interesa de la lista creada con split. y con expand=True hago que lo que devuelve sea un dataframe, adicional al final con [5] tomo solamente la columna de los valores de lat y long
```

```python
df
```

```python
df["precio*m2"] = df["precio"] / df["metroscuadradostotales"] #creo una nueva columna
```

```python
df.head()
```

```python
df= df.loc[(df["lat"]!="null")&(df["long"]!="null"),:]
```

```python
df["lat"]=df["lat"].astype(float)
df["long"]=df["long"].astype(float)
```

```python
df.shape
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.dtypes
```

```python
df.drop_duplicates(subset=["lat","long","precio","metroscuadradostotales"],keep="last",inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv("../output/dataset_final.csv", index=False, columns=["cuartos","cuartosdebaño","estacionamiento","metroscuadradostotales","precio","precio*m2","lat","long"])
```
