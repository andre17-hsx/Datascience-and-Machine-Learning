```python
#uso de apply
```

```python
def category(value): #primero creo la funcion que quiero hacer 
    if value > 120000:
        return "A"
    elif value > 60000:
        return "B"
    else:
        return "C"
```

```python
df["category"]= df["precio"].apply(category)
```

```python
df
```

```python
#uso de apply en dataframes
```

```python
def calculoMetricas(row):
    pmt2= row["precio"]/row["metroscuadradostotales"]
    pmt22= row["precio"]*row["metroscuadradostotales"]
    return [pmt2,pmt22]
```

```python
df[["a","b"]]= df.apply(calculoMetricas,axis=1,result_type="expand") #Con el result_type="expand", hago que me devuelva un dataframe y luego puedo conectarlas igualando a el dataframe original con los nuevos nombres de las filas dentro de una lista.
```
